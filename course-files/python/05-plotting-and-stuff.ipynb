# 05 Plotting in Seaborn Lesson

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
surveys_complete = pd.read_csv('surveys_complete.csv', index_col=0)
```

```python
surveys_complete
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False)
```

```python
sns.set(font_scale=1.5)
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False)
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5)
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5, scatter_kws={"s": 5})
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5, markers='+')
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5, scatter_kws={'alpha':0.1, 's':5})
```

```python
sns.lmplot("weight", "hindfoot_length", 
           data=surveys_complete, fit_reg=False, height=8,
           aspect=1.5, scatter_kws={'alpha':0.3,"s": 50}, 
           hue='species_id', markers='D')
```

```python
my_fig = sns.lmplot("weight", "hindfoot_length", 
                    data=surveys_complete, fit_reg=False, 
                    height=8, 
                    aspect=1.5, 
                    scatter_kws={'alpha':0.3,"s": 50}, 
                    hue='species_id', markers='D')
my_fig.set_axis_labels('Weight', 'Hindfoot Length')
```

```python
my_fig = sns.lmplot("weight", "hindfoot_length", 
                    data=surveys_complete[(surveys_complete.plot_id == 12)], 
                    fit_reg=False, 
                    height=8, 
                    aspect=1.5, 
                    scatter_kws={'alpha':0.3,"s": 50}, 
                    hue='sex', markers='D',y_jitter=0.3)
my_fig.set_axis_labels('Weight', 'Hindfoot Length')
```

```python
my_fig = sns.lmplot("weight", "hindfoot_length", data=surveys_complete[surveys_complete.plot_id == 12], 
                    fit_reg=False, height=8, aspect=1.5, scatter_kws={'alpha':0.3,"s": 200}, 
                    hue='sex', markers='8')
my_fig.set_axis_labels('Weight', 'Hindfoot Length')
```

```python
plot_dims = (14, 9)
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
sns.boxplot('species_id','hindfoot_length', 
            data=surveys_complete)
ax.set(xlabel='Species ID', ylabel='Hindfoot Length')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
g = sns.violinplot('weight','species_id', 
                   data=surveys_complete, 
                   linewidth=0.2, orient="h", cut=0)
g.set_xscale('log', basex=10)
ax.set(ylabel='Species ID', xlabel='Weight')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
g = sns.violinplot(y='weight',x='sex', 
                   data=surveys_complete[surveys_complete.species_id == 'OL'], 
                   linewidth=0.2, cut=0)
ax.set(ylabel='Sex', xlabel='Weight')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
sns.distplot(surveys_complete['weight'], color='c')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
sns.distplot(surveys_complete['weight'], 
             kde=False, color='c', 
             hist_kws=dict(edgecolor="k", linewidth=1), 
             bins=100)
```

```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
```

```python
output_notebook()
```

```python
import numpy as np
```

```python
hist, edges = np.histogram(surveys_complete['weight'], 
                           density=True, bins=100)
my_fig = figure(title="Weight",
                background_fill_color="#EBC8EB")
my_fig.quad(top=hist, bottom=0, left=edges[:-1], 
            right=edges[1:],fill_color="#036564", 
            line_color="#033649")
show(my_fig)
```
